```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from tqdm import tqdm
import dask.dataframe as dd

# graan1, zaai ui, mais
classes = ["Graan 1", "Zaai Ui", "Mais", "Knol", "Witlof", "Pioen"]

# Specify the file path
file_path = "/home/guilly/iTransformer/dataset/Netherlands/preprocessed_data.parquet"

# Read the parquet file
df = pd.read_parquet(file_path)
# df = df[df['class_name_late'].isin(classes)]
date_range = pd.date_range(start='2023-01-17', end='2023-10-29', freq='1D')
# print(date_range)
# print(df['DATUM_BEELD'].max())

# Perform one-hot encoding on the 'class_name_late' column
df_encoded = pd.get_dummies(df, columns=["zone_id", 'class_name_late'], dtype=int)
# Convert 'datum_beeld' column to datetime type
df_encoded['DATUM_BEELD'] = pd.to_datetime(df_encoded['DATUM_BEELD'])

# Group the dataframe by 'FOI_ID_LEVERANCIER'
grouped = df_encoded.groupby('FOI_ID_LEVERANCIER')

# Create an empty list to store the interpolated dataframes
interpolated_dfs = []
# display(df_encoded[df_encoded['FOI_ID_LEVERANCIER'] == 19317])

# for group_name, group_df in tqdm(grouped):
#     # Interpolate missing values
#     # display(group_df)
#     if group_df['DATUM_BEELD'].min() > date_range[0]:
#         missing_date_range = pd.date_range(start=date_range[0], end=group_df['DATUM_BEELD'].min(), freq='D')

#         group_df = group_df.reindex(missing_date_range, fill_value=0)
#         group_df = group_df.set_index('DATUM_BEELD')

#     display(group_df)
#     print(group_df['DATUM_BEELD'].max())
#     print(date_range[-1])
#     if group_df['DATUM_BEELD'].max() < date_range[-1]:
#         missing_date_range = pd.date_range(start=group_df['DATUM_BEELD'].max(), end=date_range[-1], freq='D')
#         group_df = group_df.reindex(missing_date_range, fill_value=0)
#     group_df = group_df.set_index('DATUM_BEELD')



#     interpolated_group = group_df.interpolate(method='linear')

#     # Resample to daily frequency
#     interpolated_group = interpolated_group.resample('D').asfreq()

#     # Select dates from date_range
#     interpolated_group = interpolated_group.loc[date_range]

#     interpolated_dfs.append(interpolated_group)
grouped_list = list(grouped)
half = len(grouped_list) // 2
group1 = grouped_list[:half]
group2 = grouped_list[half:]
for i,g in enumerate([group1, group2]):
    interpolated_dfs = []
    for group_name, group_df in tqdm(g):
        # Set 'DATUM_BEELD' as the index
        group_df = group_df.set_index('DATUM_BEELD')

        # Resample to daily frequency
        interpolated_group = group_df.resample('D').asfreq()
        # Interpolate missing values
        interpolated_group = interpolated_group.interpolate(method='linear')

        # Check if there are missing dates at the start
        if interpolated_group.index.min() > date_range[0]:
            missing_date_range_start = pd.date_range(start=date_range[0], end=interpolated_group.index.min() - pd.Timedelta(days=1), freq='D')
            missing_df_start = pd.DataFrame(index=missing_date_range_start)
            missing_df_start['FOI_ID_LEVERANCIER'] = group_name  # Add the group name to the missing rows
            interpolated_group = pd.concat([missing_df_start, interpolated_group])

        # Check if there are missing dates at the end
        if interpolated_group.index.max() < date_range[-1]:
            missing_date_range_end = pd.date_range(start=interpolated_group.index.max() + pd.Timedelta(days=1), end=date_range[-1], freq='D')
            missing_df_end = pd.DataFrame(index=missing_date_range_end)
            missing_df_end['FOI_ID_LEVERANCIER'] = group_name  # Add the group name to the missing rows
            interpolated_group = pd.concat([interpolated_group, missing_df_end])

        interpolated_group.fillna(0, inplace=True)
        # Select dates from date_range
        # interpolated_group = interpolated_group.loc[date_range]

        interpolated_dfs.append(interpolated_group)

    # Concatenate the interpolated dataframes back into a single dataframe
    if i == 0:
        interpolated_df1 = pd.concat(interpolated_dfs)
    else:
        interpolated_df2 = pd.concat(interpolated_dfs)

```

```python
def preprocess(merged_df, path):
    lazy_data = pl.scan_csv(path ,separator=';')
    col_to_use = ["FOI_ID_LEVERANCIER", "date", "TYPE_WAARDE", "BEREKENDE_WAARDE"]

# Perform the left join operation
    merged_df = merged_df.join(lazy_data.select(pl.col(["FOI_ID_LEVERANCIER", "date", "TYPE_WAARDE", "BEREKENDE_WAARDE"])),
                                on=["FOI_ID_LEVERANCIER", "date", "TYPE_WAARDE"], how="outer").drop(["FOI_ID_LEVERANCIER_right", "DATUM_BEELD_right", "TYPE_WAARDE_right"])

    name = lazy_data.select(pl.first("SOORT_SIGNAAL")).collect()[0][0]
    merged_df = merged_df.rename({"BEREKENDE_WAARDE": "BEREKENDE_WAARDE_" +str(name.item())})
    # merged_df.headc ().collect()
    return merged_df

merged_df = pl.scan_csv('/home/guilly/iTransformer/dataset/Netherlands/Netherlands_interpolated.csv')
path = "/home/guilly/iTransformer/dataset/Netherlands/S1_VH_2023.csv"
merged_df = preprocess(merged_df, path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
indi_dict = {}
df_nop_groups = df_nop.groupby('FOI_ID_LEVERANCIER')
for group, values in tqdm(df_nop_groups):
    # print(group)
    cols = values.columns[target]
    one_hot = values[cols]
    # print(values.columns[target])
    # Assuming your DataFrame is named df
    class_name = one_hot.columns[one_hot.eq(1).any()][0]
    if class_name in indi_dict:
        indi_dict[class_name] += 1
    else:
        indi_dict[class_name] = 1
sorted_dict = dict(sorted(indi_dict.items()))

# Get the class names and their corresponding counts
class_names = list(sorted_dict.keys())
counts = list(sorted_dict.values())

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(class_names, counts)
plt.xticks(rotation=90, ha='right')
plt.xlabel('Class Name')
plt.ylabel('Count')
plt.title('Class Name Counts')
plt.tight_layout()
plt.show()
```
